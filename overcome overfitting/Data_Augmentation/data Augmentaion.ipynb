```python
from keras.preprocessing import image
from keras.utils import load_img,img_to_array
from keras.preprocessing.image import ImageDataGenerator
```

```python
img=load_img('dog.jpg',target_size=(200,200))
img
```

```python
genData=ImageDataGenerator(rotation_range=40,
                           width_shift_range=0.2,
                           height_shift_range=0.2,
                           horizontal_flip=True,
                           fill_mode='nearest',
                           shear_range=0.2,
                           rescale=1.0/255,
                           zoom_range=0.2)
```

```python
img=img_to_array(img)
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
i=0
for output in genData.flow(input_batch,batch_size=32,save_to_dir='aug_data'):
    i+=1
    if i==10:
        break
```
